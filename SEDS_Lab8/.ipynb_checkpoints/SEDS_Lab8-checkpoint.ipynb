#  Bayes' law
Use Bayes’ law to calculate the probability of getting a data science job if you’ve gotten an interview for the job. This could be written P(get the DS job | interview). You’ll have to use Bayesian probability methods (your intuition or beliefs) to assign values to the different components of Bayes’ law.

`Hint:`
* Usually the probability of getting a job from applying is very low, around 1%. This is based on experience and reading about how most people apply to 100 jobs in order to get 1 job. 
* P(interview | get job) is 99%. Almost no one will hire someone they didn't interview. 
* P(interview | !get job) will be set to 10%. We might interview 10% of candidates, and only one gets the job.

```python
p = 0.01 
not_p = 1 - p
p_interview_job = 0.99
p_interview_not_job = 0.1
result = (p_interview_job * p) / (p_interview_job * p + p_interview_not_job * not_p)
print(f'P(get the DS job | interview) = {result}')
```

```python
import numpy as np
import scipy.stats as stats
import seaborn as sns

# Fit different distributions to the data (e.g., Normal, Log-Normal, Gamma)
distributions = ['norm', 'lognorm', 'gamma', 'beta']
best_fit = None
best_aic = np.inf
best_params = None

# Store the fitted distributions and parameters
fitted_distributions = {}

for dist_name in distributions:
    dist = getattr(stats, dist_name)
    
    # Fit the distribution to the data
    params = dist.fit(df['MWh'])
    
    # Calculate the AIC for this fit
    log_likelihood = np.sum(np.log(dist.pdf(df['MWh'], *params)))
    aic = 2 * len(params) - 2 * log_likelihood
    
    fitted_distributions[dist_name] = {
        'params': params,
        'aic': aic
    }
    
    if aic < best_aic:
        best_aic = aic
        best_fit = dist_name
        best_params = params

# Show the best fit distribution
print(f"Best Fit Distribution: {best_fit}")
print(f"Parameters: {best_params}")
```

```python
# Plot the best fit distribution
plt.hist(df['MWh'], bins=30, edgecolor='black', density=True, alpha=0.6, label="Observed Data")
x = np.linspace(min(df['MWh']), max(df['MWh']), 100)
best_dist = getattr(stats, best_fit)
pdf_fitted = best_dist.pdf(x, *best_params)

plt.plot(x, pdf_fitted, label=f'Fitted {best_fit} Distribution', color='red')
plt.title(f'Best Fit Distribution: {best_fit}')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.legend()
plt.show()
```
